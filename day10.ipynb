```python
neww_df = pd.read_csv("1980861c6/bike_share.csv")
neww_df
```

```python
sns.regplot(data=neww_df, x='temp', y='total_rentals', marker='+')
plt.show()
```

```python
sns.residplot(data=neww_df, x='temp', y='total_rentals')
plt.show()
```

```python
sns.regplot(data=neww_df, x='temp', y='total_rentals', order=2)
plt.show()
```

```python
import numpy as np
sns.regplot(data=neww_df, x='mnth', y='total_rentals',x_estimator=np.mean, order=2)
```

```python
sns.regplot(data=neww_df,x='temp',y='total_rentals', x_bins=4, order=2)
plt.show()
```

```python
df_crosstab= pd.crosstab(neww_df["mnth"], neww_df["weekday"], values=neww_df["total_rentals"],aggfunc="mean").round(0)
```

```python
sns.heatmap(pd.crosstab(neww_df["mnth"], neww_df["weekday"],values=neww_df["total_rentals"], aggfunc="mean"))
plt.show()
```

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=False, linewidths=.5)
plt.show()
```

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=True,center=df_crosstab.loc[9, 6],linewidths=0)
plt.show()
```
